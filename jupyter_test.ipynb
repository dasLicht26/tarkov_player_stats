```python
import plotly.graph_objs as go
import numpy as np

# Beispiel-Daten
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)

# Funktion zum Erstellen der Daten für einen bestimmten Zeitpunkt
def create_trace(t_shift):
    return [
        go.Scatter(x=t, y=np.sin(t + t_shift), mode='lines', name='X(t)'),
        go.Scatter(x=t, y=np.cos(t + t_shift), mode='lines', name='Y(t)')
    ]

# Erstellen der ersten Datenreihe
data = create_trace(0)

# Erstellen des Layouts mit Slider
layout = go.Layout(
    title="Interaktiver Graph mit Slider",
    sliders=[{
        'active': 0,
        'currentvalue': {"prefix": "t-Verschiebung: "},
        'steps': [
            {'label': f'{i}', 'method': 'animate', 'args': [[f'{i}'], {'frame': {'duration': 0, 'redraw': True}}]} 
            for i in np.arange(-5, 5, 0.1)
        ]
    }]
)

# Frames für die Animation
frames = [
    go.Frame(data=create_trace(t_shift), name=f'{t_shift}') 
    for t_shift in np.arange(-5, 5, 0.1)
]

# Erstellen der Figur
fig = go.Figure(data=data, layout=layout, frames=frames)

# Anzeige des Graphen
fig.show()
```
